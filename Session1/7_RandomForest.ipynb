## Import Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

## Preparations for Prediction Models

```python
# Read the file dataset processed_data.csv
data = pd.read_csv('./dataset/archive/processed_data.csv')

data
```

## Preparations for Prediction Models

```python
data_copy = data.copy()

education_mapping = {
    'Basic': 0,
    'Bachelor': 1,
    'Master': 2,
    'PhD': 3
}
data['Education'] = data['Education'].map(education_mapping)
# Change Marital_Status to dummies
data = pd.get_dummies(data)
data.info()
```

## Split data

```python
X = data_copy[['Kidhome','Teenhome', 'Income', 'MntFishProducts',
               'MntMeatProducts', 'MntFruits','MntSweetProducts', 'MntWines',
               'MntGoldProds', 'NumDealsPurchases','NumCatalogPurchases', 
               'NumStorePurchases', 'NumWebPurchases','NumWebVisitsMonth', 
               'Recency']]
y = data_copy[['Response']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
rfc = RandomForestClassifier(random_state=42, n_estimators=100)
rfc.fit(X_train, y_train)

y_train_pred = rfc.predict(X_train)
y_pred = rfc.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
print(cm)
print(report)
```
